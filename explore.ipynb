```python
# Create a new dataframe with only non matches between roberta-base_prediction and label
df_roberta_base_not_equal_to_label = df[df['roberta-base_prediction'] != df['label']]
```

```python
# Get number of rows in df_roberta_base_not_equal_to_label
df_roberta_base_not_equal_to_label.shape[0]
```

```python
# Drop all rows where one title is null (i.e. where one title is missing) or with less than 5 characters
df_roberta_base_not_equal_to_label = df_roberta_base_not_equal_to_label.dropna(subset=['title_left', 'title_right'])
df_roberta_base_not_equal_to_label = df_roberta_base_not_equal_to_label[df_roberta_base_not_equal_to_label['title_left'].str.len() > 5]
df_roberta_base_not_equal_to_label.shape[0]
```

```python
# Drop all the rows where the titles are the same
df_roberta_base_not_equal_to_label = df_roberta_base_not_equal_to_label[df_roberta_base_not_equal_to_label['title_left'] != df_roberta_base_not_equal_to_label['title_right']]
df_roberta_base_not_equal_to_label.shape[0]
```

```python
# Randomly sample 100 rows from df_roberta_base_not_equal_to_label
df_roberta_base_not_equal_to_label = df_roberta_base_not_equal_to_label.sample(n=100, random_state=1)
df_roberta_base_not_equal_to_label.shape[0]
```

```python
# Export df_roberta_base_not_equal_to_label to csv and excel
df_roberta_base_not_equal_to_label.to_csv('data/test/testset.csv')
df_roberta_base_not_equal_to_label.to_excel('data/test/testset.xlsx')
```
